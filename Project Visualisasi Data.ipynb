# Deskripsi Data

Berikut ini merupakan data kepadatan penduduk DKI Jakarta berdasarkan setiap Kelurahan, Kecamatan, dan Kabupaten pada tahun 2013.
Kepadatan penduduk adalah perbandingan dari jumlah penduduk dibagi dengan luas wilayahnya. Data yang ada akan menunjukkan luas wilayah berdasarkan Km2 (Kilometer Persegi).
Data yang digunakan dalam project ini merupakan data yang tersedia dalam website Pemerintah Provinsi DKI Jakarta yang di unduh melalui data.go.id.
Dari data kepadatan penduduk DKI Jakarta, saya akan mencoba untuk memunculkan statistik, melakukan visualisasi data serta narasi analisis sederhana.

```python
import pandas as pd

data_kepadatan = pd.read_csv('dkikepadatankelurahan2013.csv', delimiter = ',', encoding = 'ISO-8859-1')

data_kepadatan[['TAHUN','NAMA_PROVINSI','NAMA_KABUPATEN_KOTA','NAMA_KECAMATAN','NAMA_KELURAHAN','LUAS_WILAYAHKM2','KEPADATANJIWAKM2','55-59 Laki-Laki','55-59 Perempuan','60-64 Laki-Laki','65-69 Laki-Laki','65-69 Perempuan','70-74 Laki-Laki','70-74 Perempuan','>75 Laki-Laki','>75  Perempuan']]
```

# Statistik

```python
#COUNT

data_kepadatan[['TAHUN','NAMA_PROVINSI','NAMA_KABUPATEN_KOTA','NAMA_KECAMATAN','NAMA_KELURAHAN','LUAS_WILAYAHKM2','KEPADATANJIWAKM2','55-59 Laki-Laki','55-59 Perempuan','60-64 Laki-Laki','65-69 Laki-Laki','65-69 Perempuan','70-74 Laki-Laki','70-74 Perempuan','>75 Laki-Laki','>75  Perempuan']].count()
```

```python
#KEPADATAN JIWA PER KELURAHAN

count = data_kepadatan.KEPADATANJIWAKM2.count()
s = data_kepadatan.KEPADATANJIWAKM2.sum()
ma = data_kepadatan.KEPADATANJIWAKM2.max()
mi = data_kepadatan.KEPADATANJIWAKM2.min()
me = data_kepadatan.KEPADATANJIWAKM2.mean()
md = data_kepadatan.KEPADATANJIWAKM2.median()

print('Jumlah Penduduk Total :', s, ', Jumlah Penduduk Tertinggi :', ma, ', Jumlah Penduduk Terendah :', mi, ', Rata-rata Jumlah Penduduk :', me, ', Median dari Jumlah Penduduk :', md)
```

# Visualisasi Sederhana

```python
ax = data_kepadatan.groupby('NAMA_KABUPATEN_KOTA').max()['KEPADATANJIWAKM2']
ax = ax.plot(kind='bar', title ="Jumlah Penduduk Berdasarkan Kabupaten", legend=True, fontsize=12)
ax.set_xlabel("Kabupaten", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_kepadatan.groupby('NAMA_KABUPATEN_KOTA').mean()['KEPADATANJIWAKM2']
ax = ax.plot(kind='bar', title ="Rata-rata Jumlah Penduduk Berdasarkan Kabupaten", legend=True, fontsize=12)
ax.set_xlabel("Kabupaten", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_kepadatan.groupby('NAMA_KABUPATEN_KOTA').mean()['LUAS_WILAYAHKM2']
ax = ax.plot(kind='bar', title ="Luas Wilayah Berdasarkan Kabupaten", legend=True, fontsize=12)
ax.set_xlabel("Kabupaten", fontsize=12)
ax.set_ylabel("Luas Wilayah(Km2)", fontsize=12)
```

```python
ax = data_kepadatan.groupby('NAMA_KECAMATAN').max()['KEPADATANJIWAKM2']
ax = ax.plot(kind='bar', title ="Jumlah Penduduk Berdasarkan Kecamatan", figsize = (12,4), legend=True, fontsize=12)
ax.set_xlabel("Kecamatan", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_kepadatan.groupby('NAMA_KECAMATAN').mean()['KEPADATANJIWAKM2']
ax = ax.plot(kind='bar', title ="Rata-rata Jumlah Penduduk Berdasarkan Kecamatan", figsize = (12,4), legend=True, fontsize=12)
ax.set_xlabel("Kecamatan", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_kepadatan.groupby('NAMA_KECAMATAN').max()['LUAS_WILAYAHKM2']
ax = ax.plot(kind='bar', title ="Luas Wilayah Berdasarkan Kecamatan", figsize = (12,4), legend=True, fontsize=12)
ax.set_xlabel("Kecamatan", fontsize=12)
ax.set_ylabel("Luas Wilayah(Km2)", fontsize=12)
```
